# Exercício 3 - Medidas resumo

    1) Para as variáveis numéricas, mostre: 
        a) média
        b) mediana
        c) percentis 25 e 75
        d) mínimo e máximo
    2) Para as variáveis categoricas, retorne a frequência delas

```python
#a
dados = pd.read_csv('minhaBase.csv', delimiter=';')
```

```python
dados.head()
```

```python
dados.shape
```

```python
dados.count()
```

```python
def media(coluna):
  soma = 0
  contagem = 0
  for linha in coluna:
    soma = soma + linha
    contagem += 1
  media = soma / contagem
  return media
```

```python
m2 = dados['ano'].mean()
print(f'A média do Ano é:{m2:.0f}')

m3 = dados['quilometragem'].mean()
print(f'A média da quilometragem é: {m3:.2f}')

m4 = dados['valor'].mean()
print(f'A média de Valor é: {m4:.0f}')
```

B - Mediana

```python
mediana1 = dados['valor'].median()
print(f'A mediana do valor dos veículos é de {mediana1:.2f}')

mediana2 = dados['ano'].median()
print(f'A mediana do Ano dos veiculos é de {mediana2:.2f}')

mediana3 = dados['quilometragem'].median()
print(f'A mediana de quilometragem dos veículos é de {mediana3:.2f}')
```

C - Percentis 25 e 75

```python
dados['quilometragem'].fillna(0,inplace = True)
dados.fillna(dados['quilometragem'].median())
dados['valor'].fillna(0,inplace = True)
dados.fillna(dados['valor'].median())

percentil25 = np.percentile(dados['quilometragem'],25)
print(f'P percentil de 25 da quilometragem é de {percentil25:.0f}')

percentil25 = np.percentile(dados['valor'],25)
print(f'P percentil de 25 do valor é de R${percentil25:.0f}')

percentil25 = np.percentile(dados['ano'],25)
print(f'P percentil de 25 do ano é de {percentil25:.0f}')
```

```python
percentil75 = np.percentile(dados['quilometragem'],75)
print(f'P percentil de 75 da quilometragem é de {percentil75:.0f}')

percentil75 = np.percentile(dados['valor'],75)
print(f'P percentil de 75 do valor é de R${percentil75:.0f}')

percentil75 = np.percentile(dados['ano'],75)
print(f'P percentil de 75 do ano é de {percentil75:.0f}')
```

D - Minimo e Maximo

```python
li = min(dados['ano'])
print(f'{li:.0f}')
ls = max(dados['ano'])
print(f'{ls:.0f}')
```

```python
li = min(dados['valor'])
print(li)
ls = max(dados['valor'])
print(ls)
```

```python
li = min(dados['quilometragem'])
print(li)
ls = max(dados['quilometragem'])
print(ls)
```

3/2 - Frequencia

```python
freq_nome = dados['nome'].value_counts()
print(freq_nome)

freq_zero_km = dados['zero_km'].value_counts()
print(freq_zero_km)

freq_motor = dados['motor'].value_counts()
print(freq_motor)

freq_acessorios = dados['acessórios'].value_counts()
print(freq_acessorios)
```
